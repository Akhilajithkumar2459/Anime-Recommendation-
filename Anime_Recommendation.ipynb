Creating a recommendation system

```python
def recommend_movies(genre, min_rating):
    # Filter movies based on genre and IMDb rating
    filtered_movies = df[(df['Genre'].str.contains(genre, case=False)) & (df['Rating'] >= min_rating)]
    # Sort movies by IMDb rating in descending order
    sorted_movies = filtered_movies.sort_values(by='Rating', ascending=False)
    return sorted_movies[['Anime', 'Genre', 'Rating']]
# Example: Recommend movies in the 'Action' genre with a minimum IMDb rating of 7.0
recommended_movies = recommend_movies(input("Whats the Genre: "),float(input('Whats the minimum rating: ')))
print(recommended_movies)
```
